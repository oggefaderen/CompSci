```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Exercise 1

```python
X = np.random.standard_normal(10000) * 4  # σ=4, μ=0 (default)
```

```python
# Histogram
plt.hist(X, bins=100, density=True, alpha=0.6, color='g')
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid()
plt.show()
```

```python
# Cummulative average
# Coded by hand so asymptotically O(n^2)
cum_avg = []
for i in range(len(X)):
    if len(cum_avg) == 0:
        cum_avg.append(X[i])
    else:
        cum_avg.append(np.mean(X[:i+1]))
```

```python
# Cummulative standard error
cum_se = []
for i in range(len(X)):
    if len(cum_se) == 0:
        cum_se.append(0)
    else:
        cum_se.append(np.std(X[:i+1]) / np.sqrt(i+1))
```

```python
# Distribution mean abd median
mean = np.mean(X)
median = np.median(X)
print(f'Mean: {mean:.2f}, Median: {median:.2f}')
```

```python
# Creating a figure showing the cumulative average
plt.figure(figsize=(12, 6))
plt.plot(cum_avg, label='Cumulative Average', color='blue')

# Adding error bars for the cumulative standard error
plt.fill_between(range(len(cum_avg)), np.array(cum_avg) - np.array(cum_se), np.array(cum_avg) + np.array(cum_se), color='blue', alpha=0.2, label='Standard Error')

# Adding horizontal lines for mean and median
plt.axhline(mean, color='red', linestyle='--', label='Mean')
plt.axhline(median, color='green', linestyle='--', label='Median')
```

```python
# Cummulative median
cum_median = []
for i in range(len(X)):
    if len(cum_median) == 0:
        cum_median.append(X[0])
    else:
        cum_median.append(np.median(X[:i+1]))
```

```python
# Creating figure showing cummulative median
plt.plot(cum_median, label='Cumulative Median', color='orange')
plt.title('Cumulative Average and Median with Standard Error')
plt.xlabel('Number of Observations')
plt.ylabel('Value')
plt.legend()
plt.grid()
plt.show()
```
