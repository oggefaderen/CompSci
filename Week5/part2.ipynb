### What’s the problem with random networks as a model for real-world networks according to the argument in section 3.5 (near the end)?

The core problem is that random networks predict a very narrow, bell-curve-like degree distribution centered tightly around ⟨k⟩, whereas real networks have far more variation in node degrees.

The degree distributions of real networks (the Internet, protein interaction networks, scientific collaboration networks) all show the same pattern: far more very low-degree nodes and far more very high-degree nodes than Poisson predicts, with fewer nodes clustering near the mean. The random network model concentrates too much probability mass near ⟨k⟩ and doesn't capture the "fat tails" that produce hubs in real systems.

In short, the random network model predicts a homogeneous world where everyone has roughly the same number of connections, but the real world is deeply heterogeneous.

### According to the book, why is it a problem for random networks (in terms of being a model for real-world networks that the degree-dependent clustering C(k) decreases as a function of k in real-world networks?

Random networks predict that C(k) should be independent of degree k, but real networks show C(k) decreasing with k. This is stated as a straightforward failure of the model's predictions to match observations.

The broader framing in Section 3.10 is that the random network model fails on most metrics (degree distribution, clustering coefficient, and to some extent connectedness), with the small-world property being the only thing it handles reasonably well. The decreasing C(k) is presented as one piece of that cumulative failure, evidence that real networks have some deeper order behind them that pure randomness cannot capture.
